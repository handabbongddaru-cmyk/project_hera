### 선케어 제품 구매 시 중요하게 생각하는 요소

```python
from collections import Counter

col = '선케어 제품 구매 시 중요하게 생각하는 요소'

data = df[col].dropna()

# 복수응답 분리
split_data = data.apply(lambda x: [i.strip() for i in str(x).split(',')])

# 하나로 펼치기
all_items = []
for sublist in split_data:
    all_items.extend(sublist)

# 빈도 계산
counts = Counter(all_items)

# 정렬
items = sorted(counts.items(), key=lambda x: x[1])

labels = [x[0] for x in items]
values = [x[1] for x in items]

total = len(data)
percentages = [(v / total) * 100 for v in values]

# 그래프
plt.figure()

bars = plt.barh(labels, values)

# 값 + 퍼센트 표시
for i, (v, p) in enumerate(zip(values, percentages)):
    plt.text(v + 1, i, f'{v} ({p:.1f}%)', va='center')

plt.title('선케어 제품 구매 시 중요 요소')
plt.xlabel('응답수')

plt.xlim(0, max(values) * 1.2)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Load the dataframe if it's not already loaded (this assumes df is saved from previous steps)
# If you ran previous cells, df should already be in memory.
# If running this cell independently, uncomment the line below:
# df = pd.read_csv('선케어_설문_응답_v2.csv')

# 사용할 열
cols = [
    '연령대',
    '직업 유형',
    '선케어 제품을 사용하는 상황',
    '선케어 제품 사용 빈도',
    '선케어 제품 사용 시 불편한 점',
    '선케어 제품 구매 시 중요하게 생각하는 요소',
    '톤업 선크림에 대한 의견',
    '톤업 선크림 사용 경험',
    '톤업 선크림을 사용하는 이유',
    '톤업 선크림을 주로 사용하는 상황',
    '톤업 선크림을 사용하지 않는 이유',
    '어느 정도의 톤업 효과까지 허용',
    '어떤 경우 톤업 선크림을 사용할 의향',
    '뷰티 제품을 구매하는 곳',
    '뷰티 제품 정보를 얻는 곳',
    '신뢰하는 정보 출처',
    '헤라 인지 여부',
    '헤라 브랜드 이미지'
]

# 실제 df에 존재하는 열만 선택
cols = [col for col in cols if col in df.columns]

# 결측 포함해서 문자열로 통일
heatmap_df = df[cols].copy()
for col in heatmap_df.columns:
    heatmap_df[col] = heatmap_df[col].fillna('결측').astype(str)

# Cramér's V 함수
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)

    if confusion_matrix.shape[0] < 2 or confusion_matrix.shape[1] < 2:
        return np.nan

    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape

    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)

    if min((kcorr - 1), (rcorr - 1)) <= 0:
        return np.nan

    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# 연관성 행렬 생성
n = len(cols)
cramers_matrix = pd.DataFrame(np.zeros((n, n)), index=cols, columns=cols)

for i in range(n):
    for j in range(n):
        if i == j:
            cramers_matrix.iloc[i, j] = 1.0
        else:
            cramers_matrix.iloc[i, j] = cramers_v(heatmap_df[cols[i]], heatmap_df[cols[j]])

# 히트맵
plt.figure(figsize=(16, 13))
sns.heatmap(
    cramers_matrix,
    annot=True,
    fmt='.2f',
    cmap='YlOrRd',
    vmin=0,
    vmax=1,
    square=True
)
plt.title('범주형 변수 간 연관성 히트맵 (Cramér\'s V)', fontsize=14)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
